```python
# Import dependencies
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
```

```python
# Read the data into a Pandas DataFrame. Use the `header=2` parameter when reading in the data.

contact_info_df = pd.read_excel('Resources/contacts.xlsx', header=3)
contact_info_df.head()
```

```python
# Extract the four-digit contact ID number.
contact_info_df['contact_id'] = contact_info_df['contact_info'].str.extract(r'(\d+)')
contact_info_df
```

```python
# Check the datatypes.
contact_info_df.info()
```

```python
# Convert the "contact_id" column to an int64 data type.
contact_info_df["contact_id"] = pd.to_numeric(contact_info_df["contact_id"])
contact_info_df.info()
```

```python
# Extract the name of the contact and add it to a new column.
contact_info_df["name"] = contact_info_df["contact_info"].str.extract(r'(?:"name":\s")(\w+\s\w+)')
contact_info_df
```

```python
# Extract the email from the contacts and add the values to a new column.
contact_info_df["email"] = contact_info_df["contact_info"].str.extract(r'(?:"email":\s")(.+)(?:")')
contact_info_df
```

```python
# Create a copy of the contact_info_df with the 'contact_id', 'name', 'email' columns.
cleaned_contact_df = contact_info_df[['contact_id', 'name', 'email']]
cleaned_contact_df
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 
cleaned_contact_df["first_name"] = cleaned_contact_df["name"].str.extract(r'(\w+)')
cleaned_contact_df["last_name"] = cleaned_contact_df["name"].str.extract(r'(?:\s)(\w+)')

# Drop the contact_name column
cleaned_contact_df = cleaned_contact_df[['contact_id', 'email', 'first_name', 'last_name']]
cleaned_contact_df
```

```python
# Reorder the columns
contact_info_df = cleaned_contact_df[['contact_id', 'first_name', 'last_name', 'email']]
contact_info_df
```

```python
# Check the datatypes one last time before exporting as CSV file.
contact_info_df.info()
```

```python
# Export the DataFrame as a CSV file. 
contact_info_df.to_csv("Resources/contacts.csv", encoding='utf8', index=False)
```
